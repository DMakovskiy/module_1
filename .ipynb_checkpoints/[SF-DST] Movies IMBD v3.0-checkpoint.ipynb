# 30. Какой месяц чаще всего по годам самый прибыльный?
Варианты ответа:
1. Январь
2. Июнь
3. Декабрь
4. Сентябрь
5. Май

```python
most_profit_month = {}
for i in range(int(data.release_year.min()),int(data.release_year.max())+1):
    data_year = data[data.release_year==i]
    c = Counter()
    for index, row in data_year.iterrows():
        month = row['release_date'].split('/')[0]
        c[month] += row['profit']
        most_profit_month[i] = c.most_common(1)[0][0]
d = Counter()
for i in most_profit_month.values():
    d[i] += 1
display(d.most_common())
answer_ls.append(2)
```

# 31. Названия фильмов какой студии в среднем самые длинные по количеству символов?
Варианты ответа:
1. Universal Pictures (Universal)
2. Warner Bros
3. Jim Henson Company, The
4. Paramount Pictures
5. Four By Two Productions

```python
data['title_length'] = data.original_title.map(lambda x: len(x))
def counter(movie_db, x):
    dat=pd.Series(movie_db[x].str.cat(sep='|').split('|'))
    info=dat.value_counts()
    return info
companies = counter(data,'production_companies')
for c in companies.index:
    companies[c] = data.title_length[data.production_companies.map(lambda x: True if c in x else False)].mean()
companies = pd.DataFrame(companies).sort_values(0, ascending=False)
companies
answer_ls.append(5)
```

# 35. Какие актеры чаще всего снимаются в одном фильме вместе
Варианты ответа:
1. Johnny Depp & Helena Bonham Carter
2. Hugh Jackman & Ian McKellen
3. Vin Diesel & Paul Walker
4. Adam Sandler & Kevin James
5. Daniel Radcliffe & Rupert Grint

```python
c = Counter()
for index, row in data.iterrows():
    actors = row['cast'].split('|')
    for pair in combinations(actors, 2):
        c[pair] += 1
c.most_common(5)
answer_ls.append(5)
```

# 36. У какого из режиссеров выше вероятность выпустить фильм в прибыли? (5 баллов)101
Варианты ответа:
1. Quentin Tarantino
2. Steven Soderbergh
3. Robert Rodriguez
4. Christopher Nolan
5. Clint Eastwood

```python
c = Counter()
d = Counter()
e = Counter()
for index, row in data.iterrows():
    directors = row['director'].split('|')
    for dir in directors:
        c[dir] += 1
        if row['profit'] > 0:
            d[dir] += 1
            e[dir] = d[dir] / c[dir]
e.most_common()
answer_ls.append(4)
```
